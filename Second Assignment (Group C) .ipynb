```python
def showNumbers(limit):
    for num in range((limit + 1)):
        if num %2 ==0:
            print(num, ' EVEN') 
        else:
            print(num, ' ODD')
```

```python
showNumbers(7)
```

```python
def add_multips (limit):
    add_list = [] 
    for item in range((limit+1)):
        if item %3 ==0:
            add_list.append(item)
        elif item %5 ==0:
            add_list.append(item)
    sum = 0
    for item in add_list:
        sum += item
    return(add_list, sum)
```

```python
add_multips(30)
```

```python
def unique(list):
    unique = [] 
    for item in list:
        if item not in unique:
            unique.append(item)
    return (unique)
```
