```python
#1 Write a function called showNumbers that takes a parameter called limit. 
# It should printall the numbers between 0 and limit with a label to identify the even and odd numbers.
# For example, if the limit is 3, it should print:0 EVEN 1 ODD 2 EVEN 3 ODD 

numbers = range(1, 10)
def showNumbers(limit):
  for a in range(len(limit)):
    if a % 2 == 0:
      print(a, " Even")
    else:
      print(a, " Odd")
```

```python
#2 Write a function that returns the sum of multiples of 3 and 5 between 0and limit (parameter). 
# For example, if limit is 20, it should return the sum of 3, 5, 6, 9, 10,12, 15, 18, 20.
def limit(num):
  summation = []
  for x in range(num):
    if x % 3 == 0:
      summation.append(x)
    if x % 5 == 0:
      summation.append(x)
  print(summation)
  print(sum(summation))
multiples = int(input("Enter desired limit: ")) 
```

```python
limit(multiples)
```

```python
#4 Write a function that prints all the prime numbers between 0 and limit where limit is a parameter.
def prime(limit):
  for p in range(len(limit)):
    if p > 1:
      for x in range(2, p):
        if (p % x) == 0:
          break
      else:
        print(p)
print("Prime numbers in limit are: ")
numbers = range(1, 100)
prime(numbers)
```

```python
#5 Write a program (function!) that takes a list and returns a new list 
# that contains all the elements of the first list minus all the duplicates.
def elements(lists):
  newlist = []
  for a in lists:
    if a not in newlist:
      newlist.append(a)
  return newlist
print("newlist: ")
lists = (2, 2, 3, 3, 4, 4, 5, 6, 7, 8, 8, 9, 9)
elements(lists)
```

```python
 #7 Write a program that asks the user how many Fibonnaci numbers to generate and then generates them.
 def fib (x, y):
    return x + y

number = int(input("How many Fibonacci numbers should I generate? "))

fibonacci = []
for n in range(number):
    if n in [0, 1]:
        fibonacci += [1]
    else:
        fibonacci += [fib(fibonacci[n-2], fibonacci[n-1])]

print("The first", number, "Fibonacci numbers are:", fibonacci)
```

```python
#8 Write a function that ask the user for a string and print out whether this string is apalindrome or not.
def palindrome(p):
  t = str(input("Please input a string: "))
  for a in range(len(p)):
    if t == p[::-1]:
      print("Yes")
    else:
      print("No ")
m = "Masam"
palindrome(m)
```
